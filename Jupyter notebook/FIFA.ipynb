# Comparing Algorithms

### Reading the dataset



```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = fifa_full
```

```python
print(df.head())
```

```python
df = df.fillna(df.mean())
```

```python
print(df.head())
```

```python
# Select independent and dependent variable
X = df[["RAA", "Wins", "EFscore", "Salary"]]
y = df["Value"]
```

```python
# Split the dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model1 = LinearRegression()
model1.fit(X_train, y_train)

model2 = DecisionTreeRegressor()
model2.fit(X_train, y_train)

model4 = LinearRegression()
model4.fit(X_train, y_train)

model3 = RandomForestRegressor()
model3.fit(X_train, y_train)
```

```python
pred1 = model1.predict(X_test)
pred2 = model2.predict(X_test)
pred3 = model3.predict(X_test)
pred4 = model4.predict(X_test)
```

```python
mae1 = mean_absolute_error(y_test, pred1)
mse1 = mean_squared_error(y_test, pred1)
rmse1 = np.sqrt(mse1)
r2_1 = r2_score(y_test, pred1)

mae2 = mean_absolute_error(y_test, pred2)
mse2 = mean_squared_error(y_test, pred2)
rmse2 = np.sqrt(mse2)
r2_2 = r2_score(y_test, pred2)

mae3 = mean_absolute_error(y_test, pred3)
mse3 = mean_squared_error(y_test, pred3)
rmse3 = np.sqrt(mse3)
r2_3 = r2_score(y_test, pred3)

mae4 = mean_absolute_error(y_test, pred4)
mse4 = mean_squared_error(y_test, pred4)
rmse4 = np.sqrt(mse4)
r2_4 = r2_score(y_test, pred4)

print('Model 1 - Linear Regression:')
print('MAE:', mae1)
print('MSE:', mse1)
print('RMSE:', rmse1)
print('R2 score:', r2_1)
print('------------------------------')
print('Model 2 - Decision Tree Regressor:')
print('MAE:', mae2)
print('MSE:', mse2)
print('RMSE:', rmse2)
print('R2 score:', r2_2)
print('------------------------------')

print('Model 3 - Multiple Linear Regression:')
print('MAE:', mae4)
print('MSE:', mse4)
print('RMSE:', rmse4)
print('R2 score:', r2_4)
print('------------------------------')

print('Model 4 - Random Forest Regressor:')
print('MAE:', mae3)
print('MSE:', mse3)
print('RMSE:', rmse3)
print('R2 score:', r2_3)
print('------------------------------')
```
